```python
import pandas as pd
```

```python
blocks = pd.read_csv('block_perf.txt')
files = pd.read_csv('file_perf.txt')
```

```python
block_sizes = [64,128,256,512,1024,2048,4096,8192,16384,32768,65536]
```

```python
ax = blocks.join(pd.DataFrame(block_sizes)).set_index(0).plot(logx=True, xlabel="Block size (bytes)", ylabel="Time (s)", title="Transfer time when varying block size")
ax.get_figure().savefig('blocks.png')
```

```python
file_sizes = [1,2,4,8,16,32,64,128,256,512,1024,2048]
ax = files.join(pd.DataFrame(file_sizes)).set_index(0).plot(xlabel="File size (MB)", ylabel="Time (s)", title="Transfer time when varying file size")
ax.get_figure().savefig('files.png')
```
